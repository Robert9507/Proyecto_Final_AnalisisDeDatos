### Lectura de la base de datos

```python
import pandas as pd

df = pd.read_csv("googleplaystore_Thewebminer.csv")

df1 = df.iloc[:100]

df1.drop(['link'], axis=1)

df1


```

### MONGODB

```python
from pymongo import MongoClient


client = MongoClient("mongodb://localhost:27017/")

dm = client["Rating"]

collection = dm["GooglePlay"]

df1.dropna()


#SUBIDA DE BASE DE DATOS
records = df1.to_dict('records')


for record in records:
     collection.update_many({'_id': record['id']}, {'$set': record}, upsert=True)

```
